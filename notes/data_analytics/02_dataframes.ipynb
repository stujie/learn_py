### dictionary basics

source: https://www.youtube.com/watch?v=zmdjNSmRXF4&list=PL-osiE80TeTsWmV9i9c58mdDCSskIFdDS&index=2 

```python
# define dictionary
person = {
    "first":"Stutie",
    "last":"Banerjie",
    "email":"test@gmail.com"
}
```

```python
# make dictionary into list (w/ one person)
people = {
    "first":["Stutie"],
    "last":["Banerjie"],
    "email":["test@gmail.com"]
}
```

```python
# add people to list
people = {
    "first":["Stutie", "Jane", "Shane"],
    "last":["Banerjie", "Doe", "Hollander"],
    "email":["test@gmail.com", "JaneDoe@email.com", "hockey@gmail.com"]
}
```

```python
# access email column of dictionary
people["email"]
```

### data frame basics

```python
# create dataframe from dictionary
import pandas as pd

df = pd.DataFrame(people)

df
```

```python
# access single column of dataframe(more consistent)
df['email']
# outprints same way
df.email 
```

```python
# data frame column is a series
type(df['email'])

# dataframe - rows & columns; container of multiple series
# series - rows of a single column
```

```python
# access multiple columns of a dataframe
df[['last', 'email']]
```

```python
# access rows by integer location
df.iloc[0] # first row
```

```python
df.iloc[[0,1]] # first 2 rows
```

```python
# specify columns to show
df.iloc[[0,1],2] # emails for first 2 rows
```

```python
df
```

```python
# row associated with index 0
df.loc[0]
```

```python
# row associated with index 0 and 1
df.loc[[0,1]]
```

```python
# specify columns to show
df.loc[[0,1],['email','last']]
```

### working w/ bmw csv

```python
# read in excel file into data frame titled bmw
bmw = pd.read_csv('bmw.csv')
bmw
```

```python
# select dataframe column
bmw['transmission']
```

```python
# value counts method - counts of distinct values in column
bmw['transmission'].value_counts()
```

```python
# select dataframe row
bmw.loc[[0,1,2],'transmission']
```

```python
# slicing (rows)- last value inclusive w/ loc 
bmw.loc[0:5,'transmission']
```

```python
# slicing (columns) - last value inclusive w/ loc
bmw.loc[0:9,'transmission':'mpg']
```
