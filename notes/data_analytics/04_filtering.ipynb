### set-up

```python
people = {
    "first" : ["Corey", 'Jane', 'John'],
    "last" : ["Schafer", 'Doe', 'Doe'],
    "email" : ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

### filtering

```python
# df[df['last'] == 'Doe]

filt = df['last'] == 'Doe'

df[filt]
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

### and (&)

```python
filt1 = (df['last'] == 'Doe') & (df['first'] == 'John')

df[filt1]

df.loc[filt1, 'email']
```

### or ( | )

```python
filt2 = (df['last'] == 'Schafer') | (df['first'] == 'John')

df.loc[filt2, 'email']
```

### not (~)

```python
df.loc[~filt2, 'email']
```
